```python
# ============================================================
# 1. Importación de librerías
# ============================================================
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# ============================================================
# 2. Cargar el dataset HateBR desde Hugging Face
# ============================================================
dataset = load_dataset("franciellevargas/HateBR")

# Convertimos el split principal a DataFrame
df = pd.DataFrame(dataset["train"])

# ============================================================
# 3. Explorar las columnas disponibles
# ============================================================
print("✅ Dimensiones del dataset:", df.shape)
print("🔎 Columnas disponibles:", df.columns.tolist())

# Visualizamos los primeros registros
display(df.head(5))

# ============================================================
# 4. Mantener solo las columnas relevantes para la tarea binaria
# ============================================================
# Usamos únicamente el nivel binario (label + text)
df = df[["text", "label"]]

print("\n✅ Dataset preparado con columnas:", df.columns.tolist())
print("Número total de comentarios:", len(df))

# ============================================================
# 5. Distribución de clases
# ============================================================
class_counts = df["label"].value_counts().sort_index()
class_labels = ["No ofensivo (0)", "Ofensivo (1)"]

plt.figure(figsize=(6,4))
sns.barplot(x=class_labels, y=class_counts.values, palette="pastel")
plt.title("Distribución de clases en HateBR (nivel binario)", fontsize=13)
plt.ylabel("Número de ejemplos")
plt.show()

# Porcentaje de cada clase
percentages = (class_counts / len(df) * 100).round(2)
print("\n📊 Porcentaje de clases:")
print(percentages)
```
